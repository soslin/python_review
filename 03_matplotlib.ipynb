### Display two series on the same chart by calling .plot multiple times before we call .show

```python
from random import randint
x1 = [randint(-20,190) + x for x in range(150)]
x2 = [x + randint(-7,7) for x in x1]

plt.plot(x1, c = 'green', alpha = .8)
plt.plot(x2, c = 'firebrick', alpha = .4)

plt.show()
```

### Scatter Charts

```python
plt.scatter(range(len(x1)), x1, s = 70, c = 'firebrick', alpha = .66) #s changes size of dots
plt.scatter(range(len(x2)), x2, s = 30, c = 'navy')

plt.show()
```

### Custom Line Types

Can change the type of the line on a line chart with the ls (line style) keyword argument.

- : represents a dotted line
- -- represents a dashed line

```python
import math
x = range(-8,8)
y = [math.sin(n) for n in x]
plt.scatter(x,y)
plt.plot(x, y, ls = ':') # dotted line
plt.title('An approxmimation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
x = range(-8,8)
y = [math.sin(n) for n in x]
plt.scatter(x,y)
plt.plot(x, y, ls = '-') # dashed line
plt.title('An approxmimation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

### Rotating Text and Annotations (rotation keyword)

### Adding text

Use the .text to add arbitrary text to the chart. Specify x and y coordinates for where the text should be placed. Coordinates specify where the left boundary of the text starts.

Specify the color and fontsize keyword arguments to change the presentation of the text.

```python
plt.figure(figsize = (10,7))
x = list(range(-8,5))
x.append(3.14159)
y = [math.sin(n) for n in x]

plt.scatter(x,y)
plt.plot(x, y, ls = ':')

plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.text(.8, 0.1, '($\pi$, sin($\pi$))', fontsize=18, color='firebrick')
plt.show()
```

#### Annotations can be more fine-grained with the .annotate method.

```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c = 'orange')
plt.plot(x, z, c = 'green')

plt.annotate('Intersection', xy = (0,0), xytext = (-3,5), fontsize = 18, arrowprops = {'facecolor' : 'navy'})
                  
plt.show()
```

### Figure Size and Legends

```python
x = range(-4,5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize = (14,6))

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label= '$\sin(x)$')
plt.plot(x, cos_x, c = 'orange', label = '$\cos(x)$')

plt.legend(loc = 'upper right', fontsize = 14) # location of legend
plt.title('Approximate sin and cosine curves', fontsize = 18)
plt.xlabel('$x$')

plt.show()
```

### Subplots

```python
n_rows = 1
n_cols = 2

#sample data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

#first subplot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x,y)
plt.title('x~y')

# 2nd subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x,z)
plt.title('x ~ z')

plt.show()
```

#### shortcut for rows and columns

```python
# 1st subplot
plt.subplot(121) #number of rows, columns, and active 
plt.plot(x,y)
plt.title('x ~ y')

# 2nd subplot
plt.subplot(122)
plt.plot(x,z)
plt.title('x ~ z')
plt.suptitle('Subplots Demo') #supertitle for set of subplots

plt.show()
```

```python
#multi-row subplot

# 1st subplot
plt.subplot(211) #number of rows, columns, and active 
plt.plot(x,y)
plt.title('x ~ y')

# 2nd subplot
plt.subplot(212)
plt.plot(x,z)
plt.title('x ~ z')
plt.suptitle('Multi-row Subplots Demo') #supertitle for set of subplots

plt.show()
```

### Histograms

```python
plt.figure(figsize = (10,7)) #last value centers left, not right
x = [1, 1, 2, 3, 3, 3, 4, 4, 4, 1, 5, 5, 6]
plt.hist(x, align = 'right') #change dfault alignment (see below)
plt.show()
```

```python
# Bins
plt.figure(figsize = (10,7))
x = [1, 1, 2, 3, 3, 3, 4, 4, 4, 4, 4, 1, 5, 5, 6]
plt.hist(x, bins = [-1,0,2,3,4,5,6,7]) #add bins
plt.show()
```
